```python
import csv
arquivo = open("brasil.csv")
leitor = csv.DictReader(arquivo)
for registro in leitor:
  densidade = float(registro["habitantes"])/float(registro["area"])
  print(f"O município {registro['municipio']}/{registro['estado']} possui uma densidade demográfica de {densidade} hab/km²")
```

```python
import csv
arquivo = open("brasil.csv")
leitor = csv.DictReader(arquivo)
for registro in leitor:
  densidade = float(registro["habitantes"])/float(registro["area"])
  if densidade >= 1000:
      print(f"O município {registro['municipio']}/{registro['estado']} possui uma densidade demográfica de {densidade} hab/km²")
```
