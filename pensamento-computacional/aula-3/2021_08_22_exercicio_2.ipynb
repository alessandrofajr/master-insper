```python
#Exercício: Exercício 2: links do chat

#Baixe todos os arquivos de chat das últimas aulas no ZIP em anexo (chats.zip)

#Crie um programa que:

#Lê todos os arquivos de texto das aulas e exiba na tela apenas as mensagens que possuem links (apenas a mensagem deve ser exibida; o horário e nome não devem ser exibidos)

#Opcionalmente, você pode exibir apenas os links (sem que a mensagem inteira apareça)
```

```python
arquivo = open("GMT20210726-212633_Recording.txt") #Abrir o arquivo de texto
mensagens = []  #Uma lista em branco para adicionar os itens encontrados no loop for
for linha in arquivo: #Para cada linha no arquivo…
  partes = linha.strip().split("\t") #Separar a linha em 3 a cada tab
  if len(partes) == 3:
    mensagens.append(partes[2]) #Adicionar o último item da linha na lista em branco
```

```python
link = 'https' #Checar se contém https no conteúdo, para identificar links
for i in mensagens: #Para cada item na lista mensagens…
  if link in i: #Se tiver a variável link, então imprima esse item
    print(i)
```
